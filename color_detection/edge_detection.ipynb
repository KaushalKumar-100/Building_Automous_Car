```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

Image Read

```python
img=cv2.imread("Self_Driving_Car.jpg")
cv2.imshow("Self Driving Car",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

Gray Image

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Self Driving Gray image",img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

lets's Get Sobel X  and Y

```python
x_sobel=cv2.Sobel(img_gray,cv2.CV_64F,1,0,ksize=1)
cv2.imshow("Sobel Edge Detection",x_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
y_sobel=cv2.Sobel(img_gray,cv2.CV_64F,0,1,ksize=1)
cv2.imshow("Sobel Edge Detection",y_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

Laplacian edge Detection

```python
x_laplacian=cv2.Laplacian(img_gray,cv2.CV_64F)
cv2.imshow("Laplacian Edge Detection",x_laplacian)
cv2.waitKey()
cv2.destroyAllWindows()
```

Canny Edge Decision

```python
threshold1=40
threshold2=600
canny=cv2.Canny(img_gray,threshold1,threshold2)
cv2.imshow("Canny Edge Detection",canny)
cv2.waitKey()
cv2.destroyAllWindows()
```
