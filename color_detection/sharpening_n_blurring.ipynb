```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
img=mpimg.imread("Self_Driving_Car.jpg")
plt.imshow(img)
```

```python
image=cv2.imread("Self_Driving_Car.jpg")
cv2.imshow("Self_Driving_car",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image.shape
```

```python
img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow("gray_color",img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
sharp_kernel=np.array([
                    [0,-1,0],
                    [-1,3,1],
                    [0,-1,0]
    ])
sharp_kernel2=np.array([
                    [-1,-1,-1],
                    [-1,9,-1],
                    [-1,-1,-1]
    ])


sharpened_image=cv2.filter2D(img_gray,-1,sharp_kernel)
sharpened_image2=cv2.filter2D(img_gray,-1,sharp_kernel2)
```

```python
cv2.imshow("Sharpened_image",sharpened_image2)
cv2.waitKey()
cv2.destroyAllWindows()
```

Blurring the image


```python
Blurred_kernel=np.ones((3,3))
Blurred_kernel
```

```python
Blurred_Image=cv2.filter2D(img_gray,-1,1/9*(Blurred_kernel))
cv2.imshow("Blurred_image",Blurred_Image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
Blurred_kernel2=np.ones((8,8))*1/64
Blurred_kernel2
Blurred_Image=cv2.filter2D(img_gray,-1,(Blurred_kernel2))
cv2.imshow("Blurred_image",Blurred_Image)
cv2.waitKey()
cv2.destroyAllWindows()
```
