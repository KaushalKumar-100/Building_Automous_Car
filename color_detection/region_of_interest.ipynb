```python
from matplotlib import pyplot as plt
%matplotlib inline

from matplotlib import image as ipimg
import numpy as np
import cv2
```

```python
image=cv2.imread("image_lane_c.jpg")
cv2.imshow("lane_image",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image.shape
```

```python
height,width=image.shape[:2]
plt.xticks(np.arange(0,960,20))
plt.yticks(np.arange(0,540,20))
plt.grid(True)

plt.imshow(image)
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='gray')
```

Perform Region of Interest

```python
ROI=np.array([[(0,height),(400,350),(550,350),(width,height)]],dtype=np.int32)
blank=np.zeros_like(gray_image)

mask=cv2.fillPoly(blank,ROI,255)
masked_img=cv2.bitwise_and(gray_image,mask)
```
