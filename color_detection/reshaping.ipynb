```python
from matplotlib import pyplot as plt 
%matplotlib inline 
#show  the output inside the jyputer notebook instead of opening in new window
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python

image=cv2.imread("Self_Driving_Car.jpg")
cv2.imshow("Original Image",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height,width=image.shape[:2]
width
```

perform Rotations

```python
M_rotation=cv2.getRotationMatrix2D((width/2,height/2),180,0.9)
rotated_image=cv2.warpAffine(image,M_rotation,(width,height))
cv2.imshow("Rotated_image",rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
Translation_matrix=np.float32([[1,0,200],
                             [0,1,100]])
# Translated_image=cv2.getRotationMatrix2D((width/2,height/2),180,0.9)
Translated_image=cv2.warpAffine(image,Translation_matrix,(width,height))

cv2.imshow("Translated_Image",Translated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(Translated_image)
```

```python
Resized_image=cv2.resize(image,None,fx=.75,fy=.75,interpolation=cv2.INTER_CUBIC)
cv2.imshow("Resized_image",Resized_image)
cv2.waitKey()


cv2.destroyAllWindows()
```
