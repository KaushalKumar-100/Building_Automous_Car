```python
from matplotlib import pyplot as plt
%matplotlib inline

from matplotlib import image as ipimg
import numpy as np
import cv2
```

```python
image=cv2.imread("speed_limit_sign.jpg")
cv2.imshow("speed_Limit",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height,width=image.shape[:2]
```

```python
plt.imshow(image)
```

```python
w0=int(width*0.5)
h0=int(height*0.5)


w1=int(width*1)
h1=int(height*1)

Image_cropped=image[120:550,140:550]
cv2.imshow("Cropped Image",Image_cropped)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(Image_cropped)
```

perform Dilation and Erosion operations

```python
kernel=np.ones((3,3),np.uint8)
image_erosion=cv2.erode(image,kernel,iterations=10)
image_dilation=cv2.dilate(image,kernel,iterations=10)

cv2.imshow("Input",image)
cv2.imshow("erode",image_erosion)
cv2.imshow("dilate",image_dilation)
cv2.waitKey()
cv2.destroyAllWindows()
```
