```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image as mpimg
# from skimage.feature import hog

image=cv2.imread("truckrr.png")
cv2.imshow("truck_image",image)
cv2.waitKey()
cv2.destroyAllWindows()

 
```

```python
height,width=image.shape[:2]
```

```python
plt.imshow(image,cmap='gray')
```

```python
image_g=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_g,cmap='gray')
```

```python
threshold1=20
threshold2=250
image_canny=cv2.Canny(image_g,threshold1,threshold2)

cv2.imshow("image_canny",image_canny)
cv2.waitKey()
cv2.destroyAllWindows()
```

Peform Feature Detection

```python
fast=cv2.FastFeatureDetector_create()

keypoints=fast.detect(image_g,None)
print("Number of  keypoints Detected ",len(keypoints))
image_p=cv2.drawKeypoints(image,keypoints,None,color=(255,0,0))

cv2.imshow('image_p',image_p)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
orb=cv2.ORB_create()

keypoints=orb.detect(image_g,None)
print("Number of keypoints detectedd",len(keypoints))

image_p_orb=cv2.drawKeypoints(image_g,keypoints,None,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)


cv2.imshow('image_p_orb',image_p_orb)
cv2.waitKey()
cv2.destroyAllWindows()
```
