```python
image=mpimg.imread('speed_limit_sign.jpg')
plt.xticks(np.arange(0, 600, 20))
plt.yticks(np.arange(0, 800, 20))
plt.grid(True)
plt.imshow(image)
```

```python
height,width=image.shape[:2]
```

```python
source_point=np.float32([[210,50],[460,160],[510,550],[130,500]])
destination_point=np.float32([[0,0],[width,0],[width,height],[0,height]])
```

```python
M=cv2.getPerspectiveTransform(source_point,destination_point)
warped=cv2.warpPerspective(image,M,(width,height))
cv2.imshow("perspective_image",warped)
cv2.waitKey()
cv2.destroyAllWindows()
```
