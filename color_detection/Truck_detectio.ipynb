```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image as mpimg

image=cv2.imread("Find_Truck_color.jpg")
main_image_c=image.copy()
main_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow("truck Image",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
template_image=cv2.imread("Truck_template.jpg")
cv2.imshow("truck Image",template_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
top_left=(620,150)
bottom_right=(810,350)
cv2.rectangle(image,top_left,bottom_right,(0,0,225),5)
```

```python
cv2.imshow("trucked Imaeg",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
template_image=cv2.cvtColor(template_image,cv2.COLOR_BGR2GRAY)
```

```python
w,h = template_image.shape[::-1]
w
```

Performing Template Matching


```python
result_image=cv2.matchTemplate(main_image,template_image,cv2.TM_CCOEFF_NORMED)
result_image.shape 
```

```python
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(result_image)
```

```python
max_loc
```

```python
top_left=max_loc
bottom_right=(top_left[0]+w,top_left[1]+h)
cv2.rectangle(main_image_c,top_left,bottom_right,(0,0,255),5)
cv2.imshow("truck detected",main_image_c)
cv2.waitKey()
cv2.destroyAllWindows()
```
