```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image as mpimg
 
```

```python
image=cv2.imread("Self_Driving.jpg")
cv2.imshow("Self Driving",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
```

```python
image.shape
```

Creating Pyramid (down)

```python
scaled_down_image=cv2.pyrDown(image)
scaled_down_image.shape
```

```python
scaled_down_image2=cv2.pyrDown(scaled_down_image)
scaled_down_image2.shape
```

```python
scaled_down_image3=cv2.pyrDown(scaled_down_image2)
scaled_down_image3.shape
```

```python
cv2.imshow("original_image",image)
cv2.imshow("scaled_down_image",scaled_down_image)
cv2.imshow("scaled_down_image2",scaled_down_image2)
cv2.imshow("scaled_down_image3",scaled_down_image3)
cv2.waitKey()
cv2.destroyAllWindows()
```

Image Pyramidically scaled

```python
scaled_up_image=cv2.pyrUp(image)
scaled_up_image.shape
```

```python
scaled_up_image2=cv2.pyrUp(scaled_up_image)
scaled_up_image2.shape
```

```python
scaled_up_image3=cv2.pyrUp(scaled_up_image2)
scaled_up_image3.shape
```

```python
cv2.imshow("original_image",image)
cv2.imshow("scaled_up_image",scaled_up_image)
cv2.imshow("scaled_up_image2",scaled_up_image2)
cv2.imshow("scaled_up_image3",scaled_up_image3)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
color=('b','g','r')
for i ,col in enumerate(color):
    historgram2=cv2.calcHist([image],[i],None,[256],[0,256])
    
    plt.plot(historgram2,color=col)
    plt.xlim([0,256])
    
plt.show()
```
