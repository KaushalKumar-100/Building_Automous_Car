```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image as mpimg

image=cv2.imread("chessboard.jpg")
# main_image_c=image.copy()
# main_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow("truck Image",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
corner=cv2.cornerHarris(img_gray,5,3,0.1)
plt.imshow(corner,cmap='gray')
plt.show()
```

```python
corner_dilate=cv2.dilate(corner,np.ones((8,8),np.uint8),iterations=1)
plt.imshow(corner_dilate,cmap='gray')
```

```python
image[corner_dilate>0.01*corner_dilate.max()]=[0,0,255]
cv2.imshow("image_colored",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_truck=cv2.imread("truckrr.png")
cv2.imshow("truck",image_truck)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_gray=cv2.cvtColor(image_truck,cv2.COLOR_BGR2GRAY)
```

```python
corners_truck=cv2.cornerHarris(image_gray,7,5,0.1)
corners_dilate=cv2.dilate(corners_truck,np.ones((8,8),np.uint8),iterations=1)
plt.imshow(corners_dilate,cmap='gray')
plt.show()
```

```python
corners_dilate.max()
```

```python
image_color_corners=image_truck.copy()
image_color_corners[corners_dilate>0.03*corners_dilate.max()]=[0,0,255]
```

```python
cv2.imshow("truck",image_color_corners)
cv2.waitKey()
cv2.destroyAllWindows()
```
