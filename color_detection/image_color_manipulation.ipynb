```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
img=mpimg.imread("Self_Driving_Car.jpg")
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
height=int(img.shape[0])
print('heights: ',height, 'pixels')

Width=int(img.shape[1])
print('Width: ',Width, 'pixels')
```

```python
cv2.imshow('Self_Driving_Car.jpg',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
B,G,R=img[400,400]
print(B,G,R)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Self_Driving_Car.jpg",img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_gray.shape
```

RGB to HSV

```python
hsv_img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imshow("Hsv_image",hsv_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(hsv_img[:,:,2])
plt.title("hue channel")
```

```python
B,G,R=cv2.split(img)
B.shape
```

```python
R.shape
```

```python
G.shape
```

```python
cv2.imshow("Red Channel",R)
cv2.waitKey()
cv2.destroyAllWindows()
```

creating our own 3d image out of the blue channel

```python
zeros=np.zeros(img.shape[:2],dtype='uint8')

cv2.imshow("BLue Channel ",cv2.merge([B,zeros,zeros]))
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_merged1=cv2.merge([R,G,B])
cv2.imshow("merged Image",image_merged1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_merged2=cv2.merge([B,G,R])
cv2.imshow("merged Image",image_merged2)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
