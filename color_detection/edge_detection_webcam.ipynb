```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
def LiveCamEdgeDetection_canny(image_color):
    threshold1=30
    threshold2=80
    
    img_gray=cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
    canny=cv2.Canny(img_gray,threshold1,threshold2)
    
    return canny
```

```python
def LiveCamEdgeDetection_Laplace(image_color):
    
    
    img_gray=cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
    width=img_gray.shape[0]
    Height=img_gray.shape[1]
    
    
    M_rotation=cv2.getRotationMatrix2D((width/2,Height/2),
                                       180,1)
    rotated_image=cv2.warpAffine(img_gray,M_rotation,(width,Height))
    laplace=cv2.Laplacian(rotated_image,cv2.CV_64F)
    return laplace
```

```python
def LiveCamEdgeDetection_Sobel(image_color):
    
    img_gray=cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
    x_sobel=cv2.Sobel(img_gray,cv2.CV_64F,1,0,ksize=7)
    y_sobel=cv2.Sobel(img_gray,cv2.CV_64F,0,1,ksize=7)
    
    return x_sobel
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    cv2.imshow("Live Edge Detection ", LiveCamEdgeDetection_Laplace(frame))
    cv2.imshow("WebCam Video", frame)
    
    if cv2.waitKey(2)==13:
        break
    
cap.release()

cv2.destroyAllWindows()
    
    
```
