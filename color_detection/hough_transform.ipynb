```python
from matplotlib import pyplot as plt
%matplotlib inline

from matplotlib import image as ipimg
import numpy as np
import cv2
```

```python
image=cv2.imread("horizontal_vertical.jpg")
cv2.imshow("line",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

plt.imshow(image)
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
image_canny=cv2.Canny(gray_image,50,200,apertureSize=3)
# image_canny
cv2.imshow("Canny image",image_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lines=cv2.HoughLines(image_canny,1,np.pi/180,300)
lines
```

```python
len(lines)
```

```python
lines.shape
```

```python
if lines is not None:
    for i  in range(0,len(lines)):
        rho=lines[i][0][0]
        theta=lines[i][0][1]
        
        x0=rho*np.cos(theta)
        y0=rho*np.sin(theta)
        
        a = np.cos(theta)
        b = np. sin(theta)

        x1= int(x0+1000*(-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(gray_image, (x1, y1), (x2, y2), (255, 0,0),2)
        
cv2.imshow("Hough Lines",gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
                
```
