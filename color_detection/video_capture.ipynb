```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image as mpimg
```

```python
def gray_conversion(frame):
    gray_image=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    return gray_image
```

```python
def edge_detection(frame):
    blur_frame=cv2.GaussianBlur(frame,(5,5),0)
    threshold1=10
    threshold2=200
    edge_detected=cv2.Canny(frame,threshold1,threshold2)
    # edge_detected = cv2.convertScaleAbs(edge_detected)
    return edge_detected
```

```python
def region_of_Interest(frame):
    vertical=np.array([[(0,0),(640,0),(640,540),(0,540)]])
    mask=np.zeros_like(frame)
    cv2.fillPoly(mask,vertical,255)
    masked_image=cv2.bitwise_and(frame,mask)
    return masked_image
```

```python
def hough_line_detection(frame):
    rho=2
    theta=np.pi/180
    threshold=40
    min_line_len=200
    max_line_gap=100
    lines=cv2.HoughLinesP(frame,rho,theta,threshold,np.array([]),minLineLength=min_line_len,maxLineGap=max_line_gap)
    line_image = np. zeros((frame.shape[0],frame.shape[1],3),dtype=np.uint8)
    if lines is None:
        return line_image
    
    for x1,y1,x2,y2 in lines:
        cv2.line(line_image,(x1,y1),(x2,y2),(0,0,255),10)
    return line_image
```

```python
def oribit_detector(frame):
    orb=cv2.ORB_create()

    keypoints=orb.detect(frame,None)
    # print("Number of keypoints detectedd",len(keypoints))

    image_p_orb=cv2.drawKeypoints(frame,keypoints,None,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    return image_p_orb
```

```python
def main():
    cap=cv2.VideoCapture(0)
    while True:
        ret,frame=cap.read()
        alpha=1
        beeta=1
        gamma=0
        gray_frame = gray_conversion(frame)
        # print("Gray:", gray_frame.shape, gray_frame.dtype)

        edge_detected_frame = edge_detection(gray_frame)
        orb_image=oribit_detector(gray_frame)
        # print("Edges:", edge_detected_frame.shape, edge_detected_frame.dtype)

        intersted_region = region_of_Interest(edge_detected_frame)
        # print("ROI:", intersted_region.shape, intersted_region.dtype)

        lined_image = hough_line_detection(intersted_region)
                
        image_with_line=cv2.addWeighted(frame,alpha,lined_image,beeta,gamma)
        # print(frame.shape)
        # print(frame.dtype)
        
        # cv2.imshow('real_webcam',frame)
        cv2.imshow("lined_imag",image_with_line)
        cv2.imshow('orb_image',orb_image)
        # plt.imshow(frame)
        # print(frame.shape)
        
        if cv2.waitKey(2)==13:
            break
    cap.release()
    cv2.destroyAllWindows()
    
    
    return 0

        
        
        
        
        
        
        
```

```python
main()
```
